```python
import cv2
import numpy as np
from keras.models import load_model
```

```python
def imgPath(path):
    test_img=cv2.imread(path)
    img=np.array(test_img)
    img=cv2.resize(img,(224,224))
    img=img/255
    print(img.shape)
    img=img.reshape(1,224,224,3)
    print(img.shape)
    return img
```

```python
from tensorflow.keras.models import load_model
model_path = 'model_densenet.h5'
#model_densenet dosyası için -> 0: NonFire 1:Fire
#model_yangin2 dosyası için 0: Fire 1:NonFire
model = load_model(model_path)
```

```python
path="fire.4.png"
img=imgPath(path)
prediction=model.predict(img)
pred=max(prediction[0])
```

```python
pred=np.argmax(prediction[0])
```

```python
prob=prediction[0][pred]
```

```python
prob
```

```python
prob_y="{:.2f}".format(prob)
```

```python
if pred==1:
    label="yangin"
else:
    label="normal"

print(label)
```

```python
font=cv2.FONT_HERSHEY_SIMPLEX
test_img=cv2.imread(path)
cv2.putText(test_img,label,(35,50),font,1,(255,0,0),2)
cv2.putText(test_img,prob_y,(35,80),font,1,(255,0,0),2)
cv2.imshow("sonuc",test_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
video_path="test/tes3.mp4"
cap=cv2.VideoCapture(video_path)
if not cap.isOpened():
    print("Video dosyası açılamadı.")
    exit()
while True:
    ret,frame = cap.read()


    if not ret:
        print("Video akışı sona erdi veya hata oluştu.")
        break
    img = cv2.resize(frame, (224, 224))
    img=img/255
    img=img.reshape(1,224,224,3)
    prediction=model.predict(img)
    pred=max(prediction[0])

    pred=np.argmax(prediction[0])
    prob=prediction[0][pred]

    prob_y="{:.2f}".format(prob)
    
    if pred==1:
        label="yangin"
    else:
        label="normal"
    
    font=cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(frame,label,(35,50),font,1,(255,0,0),2)
    cv2.putText(frame,prob_y,(35,80),font,1,(255,0,0),2)
    cv2.imshow("sonuc",frame)
    if cv2.waitKey(1) & 0xFF==ord("q"):
        break

cap.release()
cv2.destroyAllWindows()


    
```
